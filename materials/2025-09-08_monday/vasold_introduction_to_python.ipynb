### Dictionaries

Dictionaries, which in other programming languages are also called *Maps*, *Hashtables* or *Associative Arrays*, are one of the most important data types in Python. A dictionary is a data type which can be used to store any number of values under a given 'key'. As each key references exactly one value, every key can only exist once within a dictionary.

Let's create an empty dictionary:

````python
latin_proverbs = {}
````

Now we can fill it up with keys and values:

````python
latin_proverbs["Alea iacta est"] = "The die is cast"
latin_proverbs["Carpe diem"] = "Fish of the day"
latin_proverbs["Veni, vidi, vici"] = "I came, I saw, I conquered"
````

````python
print(latin_proverbs)
type(latin_proverbs)
````

As soon as our dictionary has entries, we can use it to get specific values. If we'd like to get the English translation of a latin proverb

````python
latin_proverbs['Alea iacta est']
````

If we use a key, which is not in the dictionary, Python will throw a *KeyError*:

````python
latin_proverbs['Errare humanum est']
````

There are several ways to circumvent this. We could use the `in` operator to check, if the key exists or we could use the `.get()` method, which will not lead to an error, but return `None` if the key does not exist:

````python
print(latin_proverbs.get('Errare humanum est'))
````

We can even define a default value instead of `None`:

````python
print(latin_proverbs.get('Errare humanum est', 'I do not know this proverb'))
````

##### Overwriting values

We can add new keys and values to a dictionary at any time. But be aware: If a key already exists in the dictionary, it the old value will be replace by the new one:

````python
# show the old value
print(latin_proverbs['Carpe diem'])

# set a new value using an existing key
latin_proverbs['Carpe diem'] = 'Seize the day'
print(latin_proverbs['Carpe diem'])
````

Again: 

- In a dictionary
each value is referenced by a key, which means each entry consists of a key and its associated value
- A key can be used only once in a dictionary (it's unique). If used a second time, the original value is overwritten
- Dictionaries are a very useful data structure, e.g. if you have to store (and retrieve very fast) a lot of values.

## Conditions (if ... elif ... else)

- In the program flow often depending on certain states (e.g. the current value of a variable or a user input) branches must be taken. This is realized by *conditions*.

- The program checks before such a "branch" whether a condition is true or false, and then takes either one way or the other.

The general form of a condition in Python (and most higher programming languages) looks like this:

~~~
if CONDITION == True:
    do something
else: # CONDITION was not True
    do something else
~~~

Where `else` can be omitted.

````python
age = int(input('Enter your age: '))

if age < 18:
    print("You are to young!")
else:
    print("Welcome")
````

Notice the indentation after the colon: All subsequent lines with the same identation are executed if the condition is true/false.

Conditions can be nested to test more than one case:

````python
age = int(input('Enter your age: '))

if age < 18:
    print("You are too young!")
else:
    if age >= 40:
        print("You are too old!")
    else:
        print("Welcome")
````

Python allows us to avoid nesting to make this better understandable by using `elif`:

````python
age = int(input('Enter your age: '))

if age < 18:
    print("You are too young!")
elif age >= 40:
    print("You are too old!")
else:
    print("Welcome")
````

## Loops (for and while)

When programming, one or more statements often need to be repeated. For example, for each element of a sequence (e.g. for each character of a string) do something. For this purpose the `for` loop is best suited

### The for loop

The `for` loop goes element by element through a data type consisting of several elements and applies the code in the *loop body* to each element. The loop automatically terminates when there are no more elements left.

In the following example, we output each character of a string by looping in a `for` loop. After the last character, the loop terminates.

````python
sentence = "Ad astra per aspera"
for character in sentence:
    print(character)
print('loop has finished')    
````

This construct (`for element in sequence`) works for all data types that are able to supply one element after the other (e.g. strings, lists, dictionaries). Here is an example für a list, wich is generated by the `split()` method (`split()`returns a list of strings):

````python
for token in sentence.split():
    print(token)
````

`for` also works on so called generator functions. These are function, which yield one value after each other. An example for this is the `range()` function, which yields all numbers from a number range, set via parameters:

````python
total_sum = 0
for num in range(1, 11):
    print(num)
    total_sum +=  num
print('---')
print(total_sum)
````

### The while loop

While the `for` loop iterates over all elements of a sequence (or, more generally, an iterable), the `while` loop checks a condition on each loop pass and terminates when that condition is no longer satisfied.

In the following example, we use a counter (`i`), which is initialized with `0`. We then check in a `while` loop if the value of `i` is less than 3.
As long as this condition is true, we execute the loop body, where we ask the user to enter a number, which then is added to the result. When the counter reaches 3 the condition is no longer true and the loop will stop. It is really important to increase the number inside the loop, otherweise we will end up in an `infinite loop`.

````python
result = 0

i = 0
while i < 3:
    user_input = int(input('Enter a number: '))
    result += user_input
    i += 1

print(result)
````

We can break out of a loop (also works for `for` loops) by using `break`. Knowing this, we can alter our program, so that it will continue asking for numbers until the user enters `=`:

````python
sum_of_inputs = 0

while True: # infinite loop
    user_input = input('Enter a number: ')
    if user_input == '=':
        break
    sum_of_inputs += int(user_input)

print(sum_of_inputs)
````

### Nested loops

You can nest two (or more - however, this is usually not recommended because it may create a huge number of loop passes) loops inside each other. This allows, for example, to combine all elements from 2 sequences. In the next example we create a multiplication table. 

*Notice*: For counters in loops the variable names ``i``, ``j`` and ``k`` have become common (which should not be used for anything else than counters!).

````python
for i in range(1, 11):
    for j in range(1, 11):
        print(f'{i} x {j} = {i * j}')
````

Nesting loops leads to exponential run time behaviour, which easily lets the number loop iterations explode. You can try it out, by executing the next cell, then increasing the `loops_per_loop` valiable stepwise (eg. 100, 200, 300, ...) and execue it again.
Ignore the `import time` for now.

````python
import time

loops_per_loop = 500 # num of loops per loop

counter = 0
start_time = time.time()
for i in range(loops_per_loop):
    for j in range(loops_per_loop):
        for k in range(loops_per_loop):
            counter += 1
print(f"{counter} iterations in {time.time() - start_time} seconds.")
````

### Reading from files

Files can be opened, read from, overwritten and appended to.

Before a file can be read from or written to, it must be opened with the `open()` function.

`open()` expects at least one argument: The name (possibly with path) of the file

Let's open our first file. For this we will use the `with` context manager, the most common way of managing files in Python:

````python
with open('proverbia_wipo.txt', encoding='utf-8') as file:
    for line in file:
        print(line)
````

# Functions

*Functions* are procedures or, if you will, subroutines that are called from the main program

<b>The advantages of using functions are</b>:

* Functions are reusable: A function, once written, can be used multiple times in a program (or even from different programs).
* Functions prevent redundancy: each functionality exists only once in the program. This makes the program easier to maintain and reduces error-proneness.
* Functions structure the code, making it easier to understand.

### Writing a function

A function definition in Python starts with the keyword `def` (many other languages use `function` instead). After the `def` follows the name of the function. This should ideally be a verb, since a function always does something. The name of the function is terminated by a pair of round brackets and a colon. This is followed by the actual function body indented, which defines what the function does.


```
def function_name([parameter 1, [parameter 2, [...]]]):
    function body
    [return value]
```


You remember the function calls we used to make without actually knowing what we were doing?

* `print()`
* `open()`
* `len()`
* ...

All these functions have be written by someone else and made available to us, so we can use them in our code.

But let's create a new function. 
We write a (naive) `count_tokens()` function, which counts the words in a text file:

````python
def count_tokens(filename, encoding='utf-8'):
    with open(filename, encoding=encoding) as f:
        text = f.read()
    return len(text.split())
````

Now we can use this function in our code:

````python
num_of_tokens = count_tokens('proverbia_wipo.txt')
print(num_of_tokens)
````

# Modules


Modules are collections of functions written by other people (or you) that you can reuse inside of your code to perform certain things (previously made solutions for specific problems). ## Why modules are useful:
* to organize larger projects into several smaller source code files - through this parts of code that belong together are stored together in a .py file - a <b> module </b>

* However, as soon as a program has more than a few hundred lines, we should split the code into several modules, also because this makes the code easier to test and maintain

## Import and use a module from the standard library
Just to demonstrate, how useful modules are, we will import the `random` module from the Python Standard Library and use a function provided by this module:

````python
import random

random_numbers = []

for i in range(1, 10001): # loop 10 000 times
    # Add a random int between 1 and 100
    random_numbers.append(random.randint(1, 1000))

# Compute the mean of all randomly generated numbers. If random works well, this should be a number close to 500.
print(sum(random_numbers) / len(random_numbers))   
````

  * *automatic summarization (extractive, abstractive)* - a process of shortening a set of data computationally, to
    create a subset (a summary) that represents the most important or relevant information within the original content
  * *POS (part-of-speech) tagging* - categorizing words in a text (corpus) in correspondence with a particular
    part of speech, depending on the definition of the word and its context (e.g. noun phrase, noun, verb, preposition...)
  * *sentiment analysis* automatic detection of sentiments, emotions, and opinions in textual data
  * *Named Entity Recognition* identification of key information in the text and classification into a set of predefined categories
    (e.g. person, place, organization)
  * *Topic Modelling* - identifying most relevant terms / concepts of a text
